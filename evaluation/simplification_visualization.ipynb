## Analyzing number of paths

```python
# number of paths
path_number_5k = pd.read_csv("/Users/sofiiashome/Documents/Studying at WU/Bachelor's Thesis/Bachelor Thesis Coding/LoopBuddy/evaluation/5k_simplification_5_50/path_number.csv", index_col=0)
path_number_10k = pd.read_csv("/Users/sofiiashome/Documents/Studying at WU/Bachelor's Thesis/Bachelor Thesis Coding/LoopBuddy/evaluation/10k_simplification_5_50/path_number.csv", index_col=0)
path_number_15k = pd.read_csv("/Users/sofiiashome/Documents/Studying at WU/Bachelor's Thesis/Bachelor Thesis Coding/LoopBuddy/evaluation/15k_simplification_5_50/path_number.csv", index_col=0)

path_number = pd.DataFrame({
        "5k": path_number_5k["Mean"],
        "10k": path_number_10k["Mean"],
        "15k": path_number_15k["Mean"]
    })
path_number.index = path_number.index.round(2)
path_number.plot.bar()
path_number.to_csv("simplification_final/path_number.csv")
```

```python
# time taken
elapsed_time_5k = pd.read_csv("/Users/sofiiashome/Documents/Studying at WU/Bachelor's Thesis/Bachelor Thesis Coding/LoopBuddy/evaluation/5k_simplification_5_50/elapsed_time.csv", index_col=0)
elapsed_time_10k = pd.read_csv("/Users/sofiiashome/Documents/Studying at WU/Bachelor's Thesis/Bachelor Thesis Coding/LoopBuddy/evaluation/10k_simplification_5_50/elapsed_time.csv", index_col=0)
elapsed_time_15k = pd.read_csv("/Users/sofiiashome/Documents/Studying at WU/Bachelor's Thesis/Bachelor Thesis Coding/LoopBuddy/evaluation/15k_simplification_5_50/elapsed_time.csv", index_col=0)

elapsed_time = pd.DataFrame({
        "5k": elapsed_time_5k["Mean"],
        "10k": elapsed_time_10k["Mean"],
        "15k": elapsed_time_15k["Mean"]
    })
elapsed_time.index = elapsed_time.index.round(2)
elapsed_time.plot.bar()
elapsed_time.to_csv("simplification_final/elapsed_time.csv")
```

```python
# joined together
time_path_number = pd.DataFrame({
    "5k_time": elapsed_time_5k["Mean"],
    "10k_time": elapsed_time_10k["Mean"],
    "15k_time": elapsed_time_15k["Mean"],
    "5k_paths": path_number_5k["Mean"],
    "10k_paths": path_number_10k["Mean"],
    "15k_paths": path_number_15k["Mean"]
})

time_path_number.index = time_path_number.index.round(2)
time_path_number.plot.bar()
time_path_number.to_csv("simplification_final/time_path_number.csv")
```
