### Now let's check how how the street_count impacts the location of points

```python
import folium

def visualize_sampled_nodes(G, number, seed, street_count, filter=True):
    """
    Visualizes sampled nodes from a graph:
    1. Static matplotlib plot (OSMnx)
    2. Interactive Leaflet map (folium, 500x500 pixels)
    """
    
    if filter:
        nodes, coords = generate_random_filtered_coordinates(number, seed, G, street_count)
    else:
        nodes, coords,street_count = generate_random_coordinates(number, seed, G)

    # --- 1. Static OSMnx plot ---
    fig, ax = ox.plot_graph(G, show=False, close=False)
    x_vals, y_vals = zip(*coords)
    ax.scatter(x_vals, y_vals, c='red', s=40, marker='o', label='Sampled Nodes')
    ax.legend()
    plt.show()

    # --- 2. Interactive Folium map (500x500) ---
    mean_lat = sum(y for x, y in coords) / len(coords)
    mean_lon = sum(x for x, y in coords) / len(coords)
    m = folium.Map(
        location=[mean_lat, mean_lon],
        zoom_start=14,
        width='1000px',
        height='1000px'
    )

    for lon, lat in coords:
        folium.CircleMarker(
            location=[lat, lon],
            radius=5,
            color='red',
            fill=True,
            fill_color='red',
            fill_opacity=0.7
        ).add_to(m)

    return m  # display in notebook
```
