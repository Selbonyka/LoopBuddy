```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Analyzing number of paths

```python
# number of paths
path_number_5k = pd.read_csv("/Users/sofiiashome/Documents/Studying at WU/Bachelor's Thesis/Bachelor Thesis Coding/LoopBuddy/evaluation/5k_simplification_5_200/path_number.csv", index_col=0)
path_number_10k = pd.read_csv("/Users/sofiiashome/Documents/Studying at WU/Bachelor's Thesis/Bachelor Thesis Coding/LoopBuddy/evaluation/10k_simplification_5_200/path_number.csv", index_col=0)
path_number_15k = pd.read_csv("/Users/sofiiashome/Documents/Studying at WU/Bachelor's Thesis/Bachelor Thesis Coding/LoopBuddy/evaluation/15k_simplification_5_200/path_number.csv", index_col=0)

path_number = pd.DataFrame({
        "5k": path_number_5k["Mean"],
        "10k": path_number_10k["Mean"],
        "15k": path_number_15k["Mean"]
    })
path_number.index = path_number.index.round(2)
path_number.plot.bar()
path_number.to_csv("simplification_final_5_200/path_number.csv")
```

```python
# time taken
elapsed_time_5k = pd.read_csv("/Users/sofiiashome/Documents/Studying at WU/Bachelor's Thesis/Bachelor Thesis Coding/LoopBuddy/evaluation/5k_simplification_5_200/elapsed_time.csv", index_col=0)
elapsed_time_10k = pd.read_csv("/Users/sofiiashome/Documents/Studying at WU/Bachelor's Thesis/Bachelor Thesis Coding/LoopBuddy/evaluation/10k_simplification_5_200/elapsed_time.csv", index_col=0)
elapsed_time_15k = pd.read_csv("/Users/sofiiashome/Documents/Studying at WU/Bachelor's Thesis/Bachelor Thesis Coding/LoopBuddy/evaluation/15k_simplification_5_200/elapsed_time.csv", index_col=0)

elapsed_time = pd.DataFrame({
        "5k": elapsed_time_5k["Mean"],
        "10k": elapsed_time_10k["Mean"],
        "15k": elapsed_time_15k["Mean"]
    })
elapsed_time.index = elapsed_time.index.round(2)
elapsed_time.plot.bar()
elapsed_time.to_csv("simplification_final_5_200/elapsed_time.csv")
```

```python
# joined together
time_path_number = pd.DataFrame({
    "5k_time": elapsed_time_5k["Mean"],
    "10k_time": elapsed_time_10k["Mean"],
    "15k_time": elapsed_time_15k["Mean"],
    "5k_paths": path_number_5k["Mean"],
    "10k_paths": path_number_10k["Mean"],
    "15k_paths": path_number_15k["Mean"]
})

time_path_number.index = time_path_number.index.round(2)
time_path_number.plot.bar()
time_path_number.to_csv("simplification_final_5_200/time_path_number.csv")
```

```python
# time taken
badness_5k = pd.read_csv("/Users/sofiiashome/Documents/Studying at WU/Bachelor's Thesis/Bachelor Thesis Coding/LoopBuddy/evaluation/5k_simplification_5_200/badness.csv", index_col=0)
badness_10k = pd.read_csv("/Users/sofiiashome/Documents/Studying at WU/Bachelor's Thesis/Bachelor Thesis Coding/LoopBuddy/evaluation/10k_simplification_5_200/badness.csv", index_col=0)
badness_15k = pd.read_csv("/Users/sofiiashome/Documents/Studying at WU/Bachelor's Thesis/Bachelor Thesis Coding/LoopBuddy/evaluation/15k_simplification_5_200/badness.csv", index_col=0)

badness = pd.DataFrame({
        "5k_avoid": badness_5k["Avoid"],
        "10k_avoid": badness_10k["Avoid"],
        "15k_avoid": badness_15k["Avoid"],
        "5k_paved": badness_5k["Paved"],
        "10k_paved": badness_10k["Paved"],
        "15k_paved": badness_15k["Paved"]
    
    })
badness.index = elapsed_time.index.round(2)
badness.plot.bar()
badness.to_csv("simplification_final_5_200/badness.csv")
```

```python
### Concatenating the time/number results
time_path_number50 = pd.read_csv("simplification_final_5_50/time_path_number.csv", index_col=0)

final_time_path_number = pd.concat([time_path_number50, time_path_number])
final_time_path_number.to_csv("final_simplification/time_path_number.csv")

### Elapsed time results
elapsed_time50 = pd.read_csv("simplification_final_5_50/elapsed_time.csv", index_col=0)

final_elapsed_time = pd.concat([elapsed_time50, elapsed_time])
final_elapsed_time.to_csv("final_simplification/elapsed_time.csv")


#Path number results
path_number50 = pd.read_csv("simplification_final_5_50/path_number.csv", index_col=0)

final_path_number = pd.concat([path_number50, path_number])
final_path_number.to_csv("final_simplification/path_number.csv")

# Badness results:
badness50 = pd.read_csv("simplification_final_5_50/badness.csv", index_col=0)

final_badness = pd.concat([badness50, badness])
final_badness.to_csv("final_simplification/badness.csv")


```
